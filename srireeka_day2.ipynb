```python
total=0
for i in range(1,6):
  price=float(input(f"Enter price of item{i}:"))
  total+=price
  print("Total Bill=",total)

```

```python
total=0
for i in range(1,6):
  mark=float(input(f"Enter mark {i}:"))
  total+=mark
  percentage=total/5
  print("percentage=",percentage,"%")
```

```python
choice=int(input("1.Celsius to Fahrenheit\n2.Fahrenheit to Celsius\nEnter choice:"))
if choice==1:
  c=float(input("Enter Celsius:"))
  f=(c*9/5)+32
  print("Fahrenheit=",f)
else:
    f=float(input("Enter Fahrenheit:"))
    c=(f-32)*5/9
    print("Celsius=",c)

```

```python
amount=float(input("Enter Recharge Amount:"))
gst=float(input("Enter GST %:"))
final=amount+(amount*gst/100)
print("Final Amount=",final)
```

```python
balance=float(input("Enter Balance:"))
withdraw=float(input("Enter Withdrawal Amount:"))
if withdraw<=balance:
  print("Withdrawal Successful")
else:
  print("Insufficient Balance")
```
